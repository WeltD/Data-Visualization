```python
import pandas as pd
```

```python
# Read the CPIH data
cpih = pd.read_csv('CPIH.csv', index_col=0, parse_dates=True)
# Remove all the cloumns with missing values
cpih = cpih.dropna(axis=1, how='all')
# Remove all the rows with missing values
cpih = cpih.dropna(axis=0, how='any')
#Transform the Date cloumns to 2012-12 format from Dec-12 format
cpih.index = pd.to_datetime(cpih.index, format='%b-%y').strftime('%Y-%m')

cpih
```

```python
#Read the Electricity data
elec = pd.read_csv('Electricity.csv', index_col=0, parse_dates=True)
# Remove all the cloumns with missing values
elec = elec.dropna(axis=1, how='all')
# Remove all the rows with missing values
elec = elec.dropna(axis=0, how='any')
#Transform the Date cloumns to 2012-12 format from Dec-12 format
elec.index = pd.to_datetime(elec.index, format='%b-%y').strftime('%Y-%m')

elec
```

```python
#Read the Gas data
gas = pd.read_csv('Gas.csv', index_col=0, parse_dates=True)
# Remove all the cloumns with missing values
gas = gas.dropna(axis=1, how='all')
# Remove all the rows with missing values
gas = gas.dropna(axis=0, how='any')
#Transform the Date cloumns to 2012-12 format from Dec-12 format
gas.index = pd.to_datetime(gas.index, format='%b-%y').strftime('%Y-%m')

gas
```

```python
#Read the Food data
food = pd.read_csv('Food.csv', index_col=0, parse_dates=True)
# Remove all the cloumns with missing values
food = food.dropna(axis=1, how='all')
# Remove all the rows with missing values
food = food.dropna(axis=0, how='any')
#Transform the Date cloumns to 2012-12 format from Dec-12 format
food.index = pd.to_datetime(food.index, format='%b-%y').strftime('%Y-%m')

food
```

```python
#Read the House data
house = pd.read_csv('House.csv', index_col=0, parse_dates=True)
# Remove all the cloumns with missing values
house = house.dropna(axis=1, how='all')
# Remove all the rows with missing values
house = house.dropna(axis=0, how='any')
#Transform the Date cloumns to 2012-12 format from Dec-12 format
house.index = pd.to_datetime(house.index, format='%b-%y').strftime('%Y-%m')

house
```

```python
#Read the Motor data
motor = pd.read_csv('Motor.csv', index_col=0, parse_dates=True)
# Remove all the cloumns with missing values
motor = motor.dropna(axis=1, how='all')
# Remove all the rows with missing values
motor = motor.dropna(axis=0, how='any')
#Transform the Date cloumns to 2012-12 format from Dec-12 format
motor.index = pd.to_datetime(motor.index, format='%b-%y').strftime('%Y-%m')

motor
```

```python
#Read the Railway data
rail = pd.read_csv('Railway.csv', index_col=0, parse_dates=True)
# Remove all the cloumns with missing values
rail = rail.dropna(axis=1, how='all')
# Remove all the rows with missing values
rail = rail.dropna(axis=0, how='any')
#Transform the Date cloumns to 2012-12 format from Dec-12 format
rail.index = pd.to_datetime(rail.index, format='%b-%y').strftime('%Y-%m')

rail
```

```python
#Read the Rental data
rent = pd.read_csv('Rental.csv', index_col=0, parse_dates=True)
# Remove all the cloumns with missing values
rent = rent.dropna(axis=1, how='all')
# Remove all the rows with missing values
rent = rent.dropna(axis=0, how='any')
#Transform the Date cloumns to 2012-12 format from Dec-12 format
rent.index = pd.to_datetime(rent.index, format='%b-%y').strftime('%Y-%m')

rent
```

```python
#Join all the dataframes together based on the Date cloumns
df = pd.concat([cpih, elec, gas, food, house, motor, rail, rent], axis=1, join='inner')
df
```

```python
# Print the Max and Min values of the data
print(df.max())
print(df.min())
```

```python
#Rename the cloumns
df.columns = ['CPIH', 'Electricity', 'Gas', 'Food', 'House', 'Fuel', 'Railway', 'Rental']
```

```python
# 将现在的index转换为cloumns，并且命名为time
df.reset_index(inplace=True)
df.rename(columns={'index':'time'}, inplace=True)

df
```

```python
# Save the data to a json file
df.to_json('data.json', orient='records')
```
